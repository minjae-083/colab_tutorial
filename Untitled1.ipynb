```python
import seaborn as sb
import matplotlib.pyplot as plt

df = sb.load_dataset('titanic')

print(df.describe())
```

```python
# 범주형 변수 분포 확인
print(df['embarked'].value_counts())
df['embarked'].value_counts().plot(kind='bar')
plt.title('Number of People per Harbor')
plt.xlabel('Harbor')
plt.ylabel('Number of People')
plt.show()
```

```python
# 관계 시각화, 산점도
plt.scatter(df['age'], df['fare'], alpha=0.5)
plt.title('Relationship between Age and Fare')
plt.xlabel('age')
plt.ylabel('fare')
plt.show()
```

```python
print(df[['age', 'fare']].corr())
```

```python
print(df.groupby('pclass')[['age', 'fare', 'survived']].mean())
```

```python
# 그룹별 생존율 시각화
df.groupby('pclass')['survived'].mean().plot(kind='bar')
plt.title('Survived Rate per Class')
plt.xlabel('Class')
plt.ylabel('Survived Rate')
plt.show()
```

```python
pivot = df.pivot_table(index='sex', columns='pclass', values='survived', aggfunc='mean')
print(pivot)
pivot.plot(kind='bar')
plt.title('Survived Rate per Sex and Class')
plt.xlabel('Sex')
plt.ylabel('Survived Rate')
```
